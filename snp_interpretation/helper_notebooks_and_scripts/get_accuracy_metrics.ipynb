```python
import pandas as pd
import numpy as np

from plotnine import *
import seaborn as sns

%matplotlib inline
```

```python
by_fold_auroc = {'cluster':[], 'fold0':[], 'fold1':[], 'fold2':[], 'fold3':[], 'fold4':[],
                 'fold5':[], 'fold6':[], 'fold7':[], 'fold8':[], 'fold9':[]}
by_fold_auprc = {'cluster':[], 'fold0':[], 'fold1':[], 'fold2':[], 'fold3':[], 'fold4':[],
                 'fold5':[], 'fold6':[], 'fold7':[], 'fold8':[], 'fold9':[]}
for cluster in range(1, 25):
    for fold in range(10):
        with open('/mnt/lab_data3/soumyak/adpd/gkmsvm/Cluster'+str(cluster)+'/fold'+str(fold)+'/test/accuracy.txt') as acc:
            lines = acc.readlines()
            assert lines[0].strip().split(' ')[1].strip(';') == str(cluster)
            assert lines[0].strip().split(' ')[3] == str(fold)
            auroc = float(lines[1].strip().split(' ')[1])
            assert lines[1].strip().split(' ')[0].strip(':') == 'AUROC'
            auprc = float(lines[2].strip().split(' ')[1])
            assert lines[2].strip().split(' ')[0].strip(':') == 'AUPRC'
            by_fold_auroc['fold'+str(fold)].append(auroc)
            by_fold_auprc['fold'+str(fold)].append(auprc)
    by_fold_auroc['cluster'].append('cluster'+str(cluster))
    by_fold_auprc['cluster'].append('cluster'+str(cluster))
            
auroc_scores = pd.DataFrame.from_dict(by_fold_auroc)           
auprc_scores = pd.DataFrame.from_dict(by_fold_auprc)
#auroc_scores.set_index('cluster', inplace=True)
#auprc_scores.set_index('cluster', inplace=True)

display(auroc_scores)
display(auprc_scores)

auroc_scores.to_csv('/mnt/lab_data3/soumyak/adpd/auroc_scores.tsv', sep='\t', index=False)
auprc_scores.to_csv('/mnt/lab_data3/soumyak/adpd/auprc_scores.tsv', sep='\t', index=False)
```

```python
molten_auprc = pd.melt(auprc_scores, id_vars=['cluster'])
molten_auroc = pd.melt(auroc_scores, id_vars=['cluster'])

molten_auprc.drop(columns=['variable'], inplace=True)
molten_auroc.drop(columns=['variable'], inplace=True)

molten_auprc.columns = ['cluster', 'auprc']
molten_auroc.columns = ['cluster', 'auroc']

clusters = [str(cluster.split('cluster')[1]) for cluster in molten_auprc.cluster[:24]]
print(clusters)

print(molten_auprc.head())
print(molten_auroc.head())
```

```python
sns.set(rc={'figure.figsize':(11,7)})
sns.set_style("whitegrid")

auprc_plot = sns.boxplot(y='auprc', x='cluster', 
                         data=molten_auprc, 
                         width=0.5)
auprc_plot.set(xticklabels=clusters, xlabel="Cluster", ylabel = "SVM auPRC")
auprc_plot.set(ylim=(0.8, 1.0))
auprc_plot.grid(True)
print(auprc_plot)
```

```python
sns.set(rc={'figure.figsize':(11,7)})
sns.set_style("whitegrid")

auroc_plot = sns.boxplot(y='auroc', x='cluster', 
                         data=molten_auroc, 
                         width=0.5)
auroc_plot.set(xticklabels=clusters, xlabel="Cluster", ylabel = "SVM auROC")
auroc_plot.set(ylim=(0.8, 1.0))
auroc_plot.grid(True)
print(auroc_plot)
```

```python
auprc_fig = auprc_plot.get_figure()
auroc_fig = auroc_plot.get_figure()

auprc_fig.savefig('/mnt/lab_data3/soumyak/adpd/auprc_plot.pdf')
auroc_fig.savefig('/mnt/lab_data3/soumyak/adpd/auroc_plot.pdf')
```
