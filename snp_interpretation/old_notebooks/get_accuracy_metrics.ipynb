```python
import pandas as pd
```

```python
by_fold_auroc = {'cluster':[], 'fold0':[], 'fold1':[], 'fold2':[], 'fold3':[], 'fold4':[],
                 'fold5':[], 'fold6':[], 'fold7':[], 'fold8':[], 'fold9':[]}
by_fold_auprc = {'cluster':[], 'fold0':[], 'fold1':[], 'fold2':[], 'fold3':[], 'fold4':[],
                 'fold5':[], 'fold6':[], 'fold7':[], 'fold8':[], 'fold9':[]}
for cluster in range(1, 25):
    for fold in range(10):
        with open('/mnt/lab_data3/soumyak/adpd/clusters_gkmsvm/Cluster'+str(cluster)+'/fold'+str(fold)+'/test/accuracy.txt') as acc:
            lines = acc.readlines()
            assert lines[0].strip().split(' ')[1].strip(';') == str(cluster)
            assert lines[0].strip().split(' ')[3] == str(fold)
            auroc = lines[1].strip().split(' ')[1]
            assert lines[1].strip().split(' ')[0].strip(':') == 'AUROC'
            auprc = lines[2].strip().split(' ')[1]
            assert lines[2].strip().split(' ')[0].strip(':') == 'AUPRC'
            by_fold_auroc['fold'+str(fold)].append(auroc)
            by_fold_auprc['fold'+str(fold)].append(auprc)
    by_fold_auroc['cluster'].append('cluster'+str(cluster))
    by_fold_auprc['cluster'].append('cluster'+str(cluster))
            
auroc_scores = pd.DataFrame.from_dict(by_fold_auroc)           
auprc_scores = pd.DataFrame.from_dict(by_fold_auprc)
auroc_scores.set_index('cluster', inplace=True)
auprc_scores.set_index('cluster', inplace=True)

display(auroc_scores)
display(auprc_scores)

auroc_scores.to_csv('/mnt/lab_data3/soumyak/adpd/auroc_scores.csv', sep='\t')
auprc_scores.to_csv('/mnt/lab_data3/soumyak/adpd/auprc_scores.csv', sep='\t')
```
